```python
import pandas as pd
df=pd.read_csv("AirPassengers.csv")
print(df.head())
print(df.tail())
df['Month']=pd.to_datetime(df['Month'],format='%Y-%m')
print(df.head())
df.index=df['Month']
del df['Month']
print(df.head())
```

```python


import matplotlib.pyplot as plt
import seaborn as sns

sns.lineplot(data=df)
plt.ylabel("no of passengers")

```

```python
import matplotlib.pyplot as plt
rolling_mean=df.rolling(7).mean()
rolling_std=df.rolling(7).std()

plt.plot(df,color="blue",label="original passanger data")

plt.plot(rolling_mean,color="red",label="rolling mean passanger number")

plt.plot(rolling_std,color="black",label="rolling standard deviation in passanger number")

plt.title("passanger time series,Rolling mean,Standard Deviation")

plt.legend(loc="best")


from statsmodels.tsa.seasonal import seasonal_decompose
decompose=seasonal_decompose(df['#Passengers'],model='additive',period=7)
decompose.plot()
plt.show()
```
